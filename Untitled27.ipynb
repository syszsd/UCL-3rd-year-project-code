```python
def trim(df, max_size, min_size, column):
    df = df.copy()
    original_class_count = len(df[column].unique())
    print('Original Number of classes in dataframe:', original_class_count)

    sample_list = []
    groups = df.groupby(column)

    for label in df[column].unique():
        group = groups.get_group(label)
        sample_count = len(group)

        # If the current number of category samples exceeds the maximum value, randomly sample to the maximum value
        if sample_count > max_size:
            samples = group.sample(n=max_size, random_state=0)
            sample_list.append(samples)
        
        # If the number of samples in the current category is greater than or equal to the minimum, keep the entire category
        elif sample_count >= min_size:
            sample_list.append(group)

    # Merge all sampled categories and reset indexes
    df = pd.concat(sample_list, axis=0).reset_index(drop=True)
    final_class_count = len(df[column].unique())

    # Check for consistency between old and new category numbers
    if final_class_count != original_class_count:
        print('Dataframe has a reduced number of classes')

    # Print out the number of samples for each category
    balance = df[column].value_counts()
    print(balance)

    return df
```

```python
def balance(df, aug_name, max_samples, min_samples, column, working_dir, image_size):
    # Copy the original data set to prevent modification of the original data
    df = df.copy()
    # Use the trim function to prune the dataset to ensure that the number of samples in each category is between the specified minimum and maximum values
    df = trim(df, max_samples, min_samples, column)    
    
    # Creating a directory to store enhanced images
    aug_dir = os.path.join(working_dir, aug_name)
    if os.path.isdir(aug_dir):
        shutil.rmtree(aug_dir)  # If the directory already exists, delete it first
    os.mkdir(aug_dir)  # Create a new enhanced image directory
    
    # Create separate catalogs for each category of enhanced images
    for label in df['labels'].unique():
        dir_path = os.path.join(aug_dir, label)    
        os.mkdir(dir_path)
    
    # Initialize the Data Enhancement Generator
    total = 0
    gen = ImageDataGenerator(horizontal_flip=True, rotation_range=20, width_shift_range=.2,
                             height_shift_range=.2, zoom_range=.2)
    
    # Grouping data sets by category
    groups = df.groupby('labels')
    
    # Perform data enhancement for each category
    for label in df['labels'].unique():
        group = groups.get_group(label)
        sample_count = len(group)
        
        # If the number of samples in a category is less than the maximum number of samples
        if sample_count < max_samples:
            aug_img_count = 0
            delta = max_samples - sample_count  # Calculate the number of images that need to be enhanced
            target_dir = os.path.join(aug_dir, label)  # Setting the destination directory for storing enhanced images
            
            # Setting up a data enhancement generator to generate images from a DataFrame
            aug_gen = gen.flow_from_dataframe(group, x_col='filepaths', y_col=None, target_size=image_size,
                                              class_mode=None, batch_size=1, shuffle=False,
                                              save_to_dir=target_dir, save_prefix='aug-', color_mode='rgb',
                                              save_format='jpg')
            
            # Generate enhanced images until the target number is reached
            while aug_img_count < delta:
                images = next(aug_gen)
                aug_img_count += len(images)
            total += aug_img_count
    
    # Output the total number of enhanced images created
    print('Total Augmented images created=', total)
    
    # If an enhanced image is created, merge its path and labels into a new DataFrame
    if total > 0:
        aug_fpaths = []
        aug_labels = []
        classlist = os.listdir(aug_dir)
        for classes in classlist:
            classpath = os.path.join(aug_dir, classes)
            flist = os.listdir(classpath)
            for f in flist:
                fpath = os.path.join(classpath, f)
                aug_fpaths.append(fpath)
                aug_labels.append(classes)
        Fseries = pd.Series(aug_fpaths, name='filepaths')
        Lseries = pd.Series(aug_labels, name='labels')
        aug_df = pd.concat([Fseries, Lseries], axis=1)
        df = pd.concat([df, aug_df], axis=0).reset_index(drop=True)
    
    # Print and return the number of samples for each category in the balanced dataset
    print(list(df['labels'].value_counts()))
    return df
```

```python
def preprocess(sdir, trsplit, vsplit):
    # Define data classification
    # categories = ['train', 'test', 'val']
    filepaths = []
    labels = []

    # Iterate over the defined data categories
    classlist = os.listdir(sdir)
    for classes in classlist:  # Iterate through subdirectories (categories)
        classpath = os.path.join(sdir, classes)  # The full path of the splicing category
        flist = os.listdir(classpath)  # Get a list of all files in the category directory
        for f in flist:  # Iterate through the list of files
            fpath = os.path.join(classpath, f)  # Splice into the full path of the file
            filepaths.append(fpath)  # Adding file paths to the list
            labels.append(classes)  # Add the file's corresponding tag (category) to the list

    # Convert file paths and tag lists to pandas sequences
    Fseries = pd.Series(filepaths, name='filepaths')
    Lseries = pd.Series(labels, name='labels')

    # Combine file paths and tag sequences into a single DataFrame
    df = pd.concat([Fseries, Lseries], axis=1)

    # Use train_test_split to split the DataFrame by a specified ratio to create the training set
    train_df, dummy_df = train_test_split(df, train_size=trsplit, shuffle=True, random_state=0, stratify=df['labels'])

    # Calculate proportion of the remaining dataset (dummy_df) to create the validation and test set
    dsplit = vsplit / (1 - trsplit)

    # Use train_test_split again to split the DataFrame to validation set and test set
    valid_df, test_df = train_test_split(dummy_df, train_size=dsplit, shuffle=True, random_state=0, stratify=dummy_df['labels'])

    # Print out the length of each dataset to verify that the segmentation is correct
    print('train_df length:', len(train_df), ' test_df length:', len(test_df), ' valid_df length:', len(valid_df))

    # Check the number of categories in each dataset to ensure that the number of categories is the same
    trcount = len(train_df['labels'].unique())
    tecount = len(test_df['labels'].unique())
    vcount = len(valid_df['labels'].unique())

    # print a warning message if the number of categories in the training set is less than the number of categories in the test set
    if trcount < tecount:
        print('number of classes in training set is less than the number of classes in test set')
        print('This will throw an error in either model.evaluate or model.predict')

    # If the number of categories in the training set and validation set is different, print a warning message and allow the user to choose whether to continue executing the program.
    if trcount != vcount:
        print('number of classes in training set not equal to number of classes in validation set')
        print('this will throw an error in model.fit')
        print('train df class count:', trcount, 'test df class count:', tecount, 'valid df class count:', vcount)
        ans = input('Enter C to continue execution or H to halt execution')
        if ans.upper() == 'H':
            print('Halting Execution')
            import sys
            sys.exit('program halted by user')

    # Print the number of images for each category in the training set to evaluate whether the categories are balanced
    print('Below is image count per class to evaluate train_df balance')
    print(list(train_df['labels'].value_counts()))
    return train_df, test_df, valid_df
```

```python
def show_image_samples(gen ):
    t_dict=gen.class_indices
    classes=list(t_dict.keys())    
    images,labels=next(gen) # get a sample batch from the generator 
    plt.figure(figsize=(20, 20))
    length=len(labels)
    if length<9:   #show maximum of 25 images
        r=length
    else:
        r=9
    for i in range(r):
        plt.subplot(3, 3, i + 1)
        image=images[i]/255
        plt.imshow(image)
        index=np.argmax(labels[i])
        class_name=classes[index]
        plt.title(class_name, color='blue', fontsize=28)
        plt.axis('off')
    plt.show()
```

```python
def print_info(test_gen, preds, print_code, save_dir, subject):
    class_dict = test_gen.class_indices  # Get class index dictionary from test generator
    labels = test_gen.labels  # Get the real tag list
    file_names = test_gen.filenames  # Get a list of test set file names
    error_list = []  # Initialize a list to collect misclassified files
    true_class = []  # Initialize a list to collect the correct category names
    pred_class = []  # Initialize a list to collect predicted category names
    prob_list = []  # Initialize a list to collect predicted probabilities
    new_dict = {}  # Initialize a new dictionary to store the inverse mapping of class indexes
    error_indices = []  # Initialize a list to collect misclassified indices
    y_pred = []  # Initialize a list to collect predicted category indices

    # Create an inverse mapping of class indexes (from integers to class names)
    for key, value in class_dict.items():
        new_dict[value] = key

    # Convert a list of class names into a list of strings
    classes = list(new_dict.values())

    errors = 0  # Initialization error count
    # Traverse the prediction results
    for i, p in enumerate(preds):
        pred_index = np.argmax(p)  # Get the predicted category index
        true_index = labels[i]  # Get the real category index
        # If the prediction is inconsistent with the reality, an error message is logged
        if pred_index != true_index:
            error_list.append(file_names[i])
            true_class.append(new_dict[true_index])
            pred_class.append(new_dict[pred_index])
            prob_list.append(p[pred_index])
            error_indices.append(true_index)
            errors += 1  # Increase error count by one
        y_pred.append(pred_index)  # Add predicted category index to list

    # Calculate total number of tests and accuracy
    tests = len(preds)
    acc = (1 - errors / tests) * 100

    # Print accuracy information
    msg = f'There were {errors} errors in {tests} test cases Model accuracy= {acc: 6.2f} %'
    print(msg)

    # If print_code is not 0 and there is an error, print error details
    if print_code != 0:
        if errors > 0:
            r = min(errors, print_code)  # Determine the number of printed error messages
            # Print header
            print(f'{"Filename"} {"Predicted Class"} {"True Class"} {"Probability"}')
            # Print details of each error
            for i in range(r):
                # Get file path from error list and split to get file name
                split1 = os.path.split(error_list[i])
                split2 = os.path.split(split1[0])
                fname = split2[1] + '/' + split1[1]
                # Format and print error messages
                print(f'{fname}{pred_class[i]}{true_class[i]}{" "}{prob_list[i]}')

        else:
            # If there are no errors, print information with 100% accuracy.
            print('No errors')

    # If there are errors, plot a bar chart of the error distribution
    if errors > 0:
        plot_bar = []  
        plot_class = []  
        for key, value in new_dict.items():        
            count = error_indices.count(key) 
            if count != 0:
                plot_bar.append(count)  
                plot_class.append(value)   
        fig = plt.figure()
        fig.set_figheight(len(plot_class) / 3)
        fig.set_figwidth(10)
        plt.style.use('fivethirtyeight')
        for i in range(0, len(plot_class)):
            c = plot_class[i]
            x = plot_bar[i]
            plt.barh(c, x)
            plt.title('Errors by Class on Test Set')
    
    # Convert to NumPy array to calculate confusion matrix
    y_true = np.array(labels)        
    y_pred = np.array(y_pred)
    
    # If the number of categories is not large, draw a confusion matrix
    if len(classes) <= 30:
        cm = confusion_matrix(y_true, y_pred)        
        length = len(classes)
        fig_width = 8 if length < 8 else int(length * .5)
        fig_height = 8 if length < 8 else int(length * .5)
        plt.figure(figsize=(fig_width, fig_height))
        sns.heatmap(cm, annot=True, vmin=0, fmt='g', cmap='Blues', cbar=False)       
        plt.xticks(np.arange(length) + .5, classes, rotation=90)
        plt.yticks(np.arange(length) + .5, classes, rotation=0)
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.title("Confusion Matrix", fontsize=34)
        plt.show()
    
    # Print classification report
    clr = classification_report(y_true, y_pred, target_names=classes, digits=4)
    print("Classification Report:\n----------------------\n", clr)
    
    # return accuracy
    return acc / 100
```

```python
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    epochs = range(len(acc))
    
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, 'bo-', label='Training accuracy')
    plt.plot(epochs, val_acc, 'go-', label='Validation accuracy')
    plt.title('Training and validation accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, 'bo-', label='Training loss')
    plt.plot(epochs, val_loss, 'go-', label='Validation loss')
    plt.title('Training and validation loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```

```python
def saver(save_path, model, model_name, subject, accuracy, img_size, scalar, offset, generator):
    # First save the model
    # Construct filename from model name, topic, and precision to two decimal places
    save_id = str(model_name + '-' + subject + '-' + str(accuracy)[:str(accuracy).rfind('.')+3] + '.h5')
    model_save_loc = os.path.join(save_path, save_id)
    model.save(model_save_loc)  # Save the model to the specified path
    print('model was saved as ' + model_save_loc)  # Print successfully saved information

    # Next create a TXT file of class names
    class_names = list(generator.class_indices.keys())
    txt_name = 'labels.txt'
    txt_save_loc = os.path.join(save_path, txt_name)
    with open(txt_save_loc, 'w') as txt_file:
        for class_name in class_names:
            txt_file.write(class_name + '\n')  # Write each class name on a new line

    print('class names txt file was saved as ' + txt_save_loc)  # Print successfully saved information

    return model_save_loc, txt_save_loc  # Returns the saving path of the model and TXT file
```

```python
sdir=r'../desktop/new5'
trsplit=.6
vsplit=.2
train_df, test_df, valid_df= preprocess(sdir, trsplit, vsplit)
max_samples_train = 300
min_samples_train = 0
max_samples_test = 100
min_samples_test = 0
img_size=(224, 224)

column='labels'
working_dir = r'../desktop/MyProject'

train_df=balance(train_df, 'aug_train', max_samples_train, min_samples_train, column, working_dir, img_size)
valid_df=balance(valid_df, 'aug_valid', max_samples_test, min_samples_test, column, working_dir, img_size)
test_df=balance(test_df, 'aug_test', max_samples_test, min_samples_test, column, working_dir, img_size)
channels=3
batch_size=60
img_shape=(img_size[0], img_size[1], channels)
length=len(test_df)
test_batch_size=sorted([int(length/n) for n in range(1,length+1) if length % n ==0 and length/n<=80],reverse=True)[0]  
test_steps=int(length/test_batch_size)
print ( 'test batch size: ' ,test_batch_size, '  test steps: ', test_steps)
```

```python
def scalar(img):    
    return img
trgen=ImageDataGenerator(preprocessing_function=scalar, horizontal_flip=True)
tvgen=ImageDataGenerator(preprocessing_function=scalar)
print('for the train generator','\r', end='') 
train_gen=trgen.flow_from_dataframe( train_df, x_col='filepaths', y_col='labels', target_size=img_size, class_mode='categorical',
                                    color_mode='rgb', shuffle=True, batch_size=batch_size)
print('for the test generator','\r', end='') 
test_gen=tvgen.flow_from_dataframe( test_df, x_col='filepaths', y_col='labels', target_size=img_size, class_mode='categorical',
                                    color_mode='rgb', shuffle=False, batch_size=test_batch_size)
print('for the valid generator','\r', end='') 
valid_gen=tvgen.flow_from_dataframe( valid_df, x_col='filepaths', y_col='labels', target_size=img_size, class_mode='categorical',
                                    color_mode='rgb', shuffle=True, batch_size=batch_size)
classes=list(train_gen.class_indices.keys())
class_count=len(classes)
train_steps=int(np.ceil(len(train_gen.labels)/batch_size))
labels=test_gen.labels
```

```python
model_name='EfficientNet-lite0'
model_url = "https://tfhub.dev/tensorflow/efficientnet/lite0/feature-vector/2"

inputs = tf.keras.Input(shape=img_shape)
base_model = hub.KerasLayer(model_url, trainable=False)(inputs)

x = BatchNormalization()(base_model)
x = Dense(256, activation='relu')(x)
x = Dropout(rate=.4)(x)
output = Dense(class_count, activation='softmax')(x)

model = Model(inputs=inputs, outputs=output)
model.compile(optimizer=Adamax(learning_rate = 0.01), loss='categorical_crossentropy', metrics=['accuracy']) 
```

```python
def testimg(sdir):
    filepaths = []
    labels = []

    # Iterate over the defined data categories
    classlist = os.listdir(sdir)
    for classes in classlist:  # Iterate through subdirectories (categories)
        classpath = os.path.join(sdir, classes)  # The full path of the splicing category
        flist = os.listdir(classpath)  # Get a list of all files in the category directory
        for f in flist:  # Iterate through the list of files
            fpath = os.path.join(classpath, f)  # Splice into the full path of the file
            filepaths.append(fpath)  # Adding file paths to the list
            labels.append(classes)  # Add the file's corresponding tag (category) to the list
    return filepaths, labels
```
